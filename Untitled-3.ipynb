```python
import numpy as np
from scipy import optimize 
import matplotlib.pyplot as plt
from scipy.stats import beta
plt.style.use('seaborn-whitegrid')
%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
# a. defining the parameters
y = 1
theta = -2
p = 0.2
x = 0.6

#b. defining the functions to be used in the function to be optimized
def u_func(z, theta = -2):
    """ calculates utility of of the agent
    
    Args:
    
        z (float): meaningless variable used just so numerical optimization is possible
        theta (integer): elasticity of utility
        
    Returns:
    
        utility function (float): utility used for the optimization
    
    """
    
    return z ** (1 + theta) / (1 + theta)

def V_insured(q, x, p, y):
    """ calculates expected utility of an insured agent
    
    Args:
    
        q (float): coverage amount
        x (float): monetary loss
        p (float): probability of of loss incursion
        y (integer) : assets
        
    Returns:
    
         expected utility of an insured agent (float): objevtive function
    
    """
    
    return p * u_func(y - x + q - p * q) + (1 - p) * u_func(y - p * q)

#c. defining the the function to be optimized 
def optimal_q(x, y, p):
    """ calculates q*
    
    Args:
    
        x (float): monetary loss
        p (float): probability of of loss incursion
        y (integer) : assets
        
    Returns:
    
         x_best (float): function used to loop over each solution for q* at each x
    
    """
    
    bounds = (0,x)
    obj = lambda q: -V_insured(q, x, p, y)
    res = optimize.minimize_scalar(obj, method='bounded', bounds=bounds)
    x_best = res.x
    return x_best
```

```python
#a. making a grid of x in the specified range [0.01,0.9]
N = 100
x_min = 0.01
x_max = 0.9

xs = np.linspace(x_min, x_max, N)
qs = np.zeros(N)

#b. looping over each solution of q* at every x
for i, x in enumerate(xs):
    qs[i] = optimal_q(x, y, p)
    
#c. plotting the solutions in the specified range
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(qs, xs)
ax.set_title(' Figure 1.1: The agents optimal insurance coverage for a given monetary loss')

ax.set_ylabel('Optimal insurance coverage')
ax.set_xlabel('Monetary loss')

plt.show()
```

On the graph we see a linear optimal insurance coverage for the agent as a function of the monetary loss to be incurred, suggesting that the agent is better off with more insurance coverage, the bigger the monetary loss is.

```python
#a. defining the parameters again
y = 1
theta = -2
p = 0.2
x = 0.6   

#b. defining the functions
def V_pi(pi, q, x, y, p):
    """ calculates the expected value of an insured agent
    
    Args:
    
        q (float): insurance coverage
        pi (float): premium
        x (float): monetary loss
        p (float): probability of of loss incursion
        y (integer) : assets
        
    Returns:
    
         the function of expected value for an insured agent (float): Obejctive function
    """
    return p * u_func(y - x + q - pi) + (1 - p) * u_func(y - pi)

#b.i. defining the expected value for an agent who is not insured
V0 = V_pi(0, 0, x, y, p)
    
def optimal_pi(q, x, y, p, V0):
    """ calculates the optimal premium (pi) for a given coverage (q)
    
    Args:
    
        q (float): insurance coverage
        x (float): monetary loss
        p (float): probability of of loss incursion
        y (integer) : assets
        V0 (float): expected value of an agent who is not insured
        
    Returns:
    
        pi_bar (float): pi_bar which makes the set of acceptable premiums for the agent
        
    """
    def obj(pi):
        """ calculates the optimal premium (pi)
    
        Args:
    
            pi (float): acceptable premiums
        
        Returns:
    
            the minimally acceptable contracts (float): an expected value as least as good as not having an insurance
            
        """
        return V_pi(pi, q, x, y, p) - V0
    x0 = 0.1
    res = optimize.root(obj, x0)
    pi_bar = res.x
    return pi_bar

#c. creating the grid of q's and pi's
qs = np.linspace(0.01, 0.6, N)
pi_bars = np.zeros(N)
pis = qs*p

#d. lopping over each q to find the corresponding pi-bar to make the set of feasible premiums
for i, q in enumerate(qs):
        pi_bars[i] = optimal_pi(q, x, y, p, V0)

#e. making the figure
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(qs, pi_bars, label = 'Acceptable premiums for the agent', color = 'blue')
ax.plot(qs, pis, label = 'Premiums for insurance company to break even', color = 'orange')
ax.set_title(' Figure 2.1: Set of feasible premiums for a given coverage')
ax.fill_between(qs, pi_bars, pis, color = 'lightgrey', alpha = 0.6, label = 'Acceptable premiums for both')

ax.set_ylabel('Premium')
ax.set_xlabel('Insurance coverage')
ax.legend()

plt.show()
```

```python
#a. Setting the variables
np.random.seed(10)
N = 10000
a, b = 2, 7
x = np.random.beta(a, b, N)

#b. creating the Monte Carlo function with the lopped beta distribution inside
def MC(y, p, gamma, pi, N):
    """ calculates the agent expected value by Monte Carlo integration
    
    Args:
    
        y (integer): assets
        p (float): probability
        gamma (float): coverage amount as a fraction of the beta distributed x
        pi (float): premium
        N (integer): number of draws
        
    Returns:
    
        the mean value of the expected value of the agent (float): x_average
    
    """
    
    xs = u_func(y - (1 - gamma) * x - pi) * x
    g_func = np.mean(xs)
    return g_func
```

```python
print(MC(y, p, 0.9, 0.2, N))
print(MC(y, p,  0.45, 0.1, N))
```

We can see from the Monte Carlo integration that the first option, that is with a gamma value of 0.9 and a pi value of 0.2, that the agent gets the highets expected value. This means that if the monetary loss is incurring at a random beta-distribution the agent would be best off picking the option with a high insurance coverage (gamma) and thereby a bigger profit for the insurance company, and therefore the agent would pick the first option.
